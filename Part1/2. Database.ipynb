```python
import os
import pandas as pd
from sqlalchemy import create_engine, Column, Integer, String, Float, MetaData, Table , DECIMAL
from sqlalchemy.orm import sessionmaker , relationship
from sqlalchemy import ForeignKey
from sqlalchemy.orm import mapper

USERNAME = 'root'
PASSWORD = #'YOUR PASSWORD'
HOST = '127.0.0.1'
PORT = 3306
DATABASE = 'database2'
engine = create_engine(f'mysql+pymysql://{USERNAME}:{PASSWORD}@{HOST}:{PORT}/{DATABASE}', echo=True)

meta = MetaData()

csv1_file = '/Users/sarayetel/Desktop/Data Science/Project1/2. Database/FinalMain.csv'
csv2_file = '/Users/sarayetel/Desktop/Data Science/Project1/2. Database/Tags.csv'
csv3_file = '/Users/sarayetel/Desktop/Data Science/Project1/2. Database/GitHubLinks.csv'
csv4_file = '/Users/sarayetel/Desktop/Data Science/Project1/2. Database/CombinedCSV.csv'

df_csv1 = pd.read_csv(csv1_file)
df_csv2 = pd.read_csv(csv2_file)
df_csv3 = pd.read_csv(csv3_file)
df_csv4 = pd.read_csv(csv4_file, sep=';')

# Define and create the CSV1, CSV2, and CSV3 tables
csv1_table = Table(
    'FinalMain',
    meta,
    Column('rank', Integer, primary_key=True),
    Column('name', String(255)),
    Column('symbol', String(255)),
    Column('market_cap', Float),
    Column('price', Float),
    Column('volume24h', Float),
    Column('main_link', String(255)),
    Column('historical_link', String(255))
)

csv2_table = Table(
    'Tags',
    meta,
    Column('rank', Integer, primary_key=True),
    Column('name', String(255)),
    Column('tags', String(255))
)

csv3_table = Table(
    'GitHubLinks',
    meta,
    Column('name', String(255), primary_key=True),
    Column('github_link', String(255))
)
csv4_table = Table(
    'CombinedCSV',
    meta,
    Column('name', String(255), primary_key=True),
    Column('time_open', String(255)),
    Column('time_close', String(255)),
    Column('time_high', String(255)),
    Column('time_low', String(255)),
    Column('open', Float),
    Column('high', Float),
    Column('low', Float),
    Column('close', Float),
    Column('volume', Float),
    Column('market_cap', Float),
    Column('timestamp', String(255))
)

meta.create_all(engine)

Session = sessionmaker(bind=engine)
session = Session()

df_csv1.to_sql('FinalMain', engine, if_exists='replace', index=True)
df_csv2.to_sql('Tags', engine, if_exists='replace', index=True)
df_csv3.to_sql('GitHubLinks', engine, if_exists='replace', index=True)
df_csv4.to_sql('CombinedCSV', engine, if_exists='replace', index=True)

session.commit()
session.close() 
```
